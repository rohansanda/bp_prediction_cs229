```python
import vitaldb

import numpy as np
import scipy.signal as ss
from scipy.ndimage import median_filter, minimum_filter1d, maximum_filter1d

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

from tqdm.auto import tqdm

# colorblind friendly palette
import matplotlib as mpl
cp = ["#172A5A", "#FF7171", "#227567", "#34BAEA", "#F9D466", ]
mpl.rcParams['axes.prop_cycle'] = mpl.cycler(color=cp)

# set default font
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = ['Open Sans', 'Arial']
plt.rcParams['figure.figsize'] = [10, 5]
```

```python
df_trks = pd.read_csv('https://api.vitaldb.net/trks')  # read track list
df_cases = pd.read_csv("https://api.vitaldb.net/cases")  # read case information
```

```python
caseids = list(
    set(df_trks[df_trks['tname'] == 'SNUADC/ART']['caseid']) &
    set(df_trks[df_trks['tname'] == 'SNUADC/ECG_II']['caseid']) &
    set(df_trks[df_trks['tname'] == 'SNUADC/PLETH']['caseid']) &
    set(df_cases[df_cases['age'] > 18]['caseid']) &
    set(df_cases[df_cases['age'] >= 18]['caseid']) & 
    set(df_cases[df_cases['weight'] >= 30]['caseid']) & 
    set(df_cases[df_cases['weight'] < 140]['caseid']) & 
    set(df_cases[df_cases['height'] >= 135]['caseid']) & 
    set(df_cases[df_cases['height'] < 200]['caseid']) & 
    set(df_cases[~df_cases['opname'].str.contains("transplant", case=False)]['caseid']) & 
    set(df_cases[~df_cases['opname'].str.contains("aneurysm", case=False)]['caseid']) & 
    set(df_cases[~df_cases['opname'].str.contains("aorto", case=False)]['caseid'])& 
    set(df_cases[df_cases['ane_type'] == 'General']['caseid'])
)
print('Total {} cases found'.format(len(caseids)))
```

```python
track_names = ['ECG_II', 'ART', 'PLETH']
vf = vitaldb.VitalFile(caseids[1], track_names)

fs = 125
data = vf.to_numpy(track_names, 1/fs)
t = np.arange(data.shape[0]) / fs

ecg = data[:,0]
art = data[:,1]
ppg = data[:,2]
```

```python
nan_indices = np.where(np.isnan(ecg))[0]
diffs = np.diff(nan_indices)
one_indices = np.where(diffs == 1)[0]

if len(diffs) == 0:
    longest_sequence = 0
else:
    lengths = np.split(np.diff(one_indices), np.where(np.diff(one_indices) != 1)[0]+1)
    lengths = [l for l in lengths if len(l) > 1]
    longest_sequence = max([len(l)+1 for l in lengths])

    
print('Longest nan length: {}'.format(longest_sequence))
```

```python
first_non_nan_index = np.argmax(~np.isnan(ecg))
num_nans = first_non_nan_index
print(num_nans)
ecg2 = ecg[first_non_nan_index:]
print(np.isnan(ecg2).sum())
```

```python
missing_values = np.isnan(ecg2)
indices = np.arange(len(ecg2))
interp = np.interp(indices, indices[~missing_values], ecg2[~missing_values])
ecg2[missing_values] = interp[missing_values]

fc = 0.5
b, a = ss.butter(2, fc / (fs / 2), 'highpass', analog=False)
filtered_ecg = ss.filtfilt(b, a, ecg2, method='pad')

#filtered_ecg[np.isnan(filtered_ecg)] = np.nanmedian(filtered_ecg)
fixed_ecg = ecg2 - filtered_ecg + np.median(filtered_ecg)
print(filtered_ecg[1:400])
```

```python
ecg_median = median_filter(ecg, int(fs*0.5))
ppg_median = median_filter(ppg, int(fs*0.5))

# Subtract median from original signal to get rid of baseline drift
ecg_corrected = ecg - ecg_median
ppg_corrected = ppg - ppg_median
```

```python
# Find and remove saturated regions using np.diff
d_ecg = np.diff(ecg_corrected)
d_ecg_abs = np.abs(d_ecg)
thresh = 0.3  # Threshold for detecting saturation
sat_idx = np.where(d_ecg_abs < thresh)[0] + 1  # Add 1 to account for diff operation
sat_mask = np.zeros_like(ecg_corrected, dtype=bool)
for idx in sat_idx:
    sat_mask[idx-10:idx+10] = True  # Set a window of 10 samples around each saturation point to True

ecg_desat = np.copy(ecg_corrected)
ecg_desat[sat_mask] = np.nan  # Replace saturated regions with NaN

# Plot original and corrected signals
plt.figure(figsize=(10, 5))
plt.subplot(2, 1, 1)
plt.plot(t, ecg_corrected, label='Original')
plt.subplot(2, 1, 2)
plt.plot(t, ecg_desat, label='Corrected')
plt.legend()
plt.show()
```

```python
def um(ta=None, tb=None):
    ta = ta if ta else t.min()
    tb = tb if tb else t.max()
    twin = (t >= ta) & (t <= tb)

    plt.figure(figsize=(12,8), dpi=100)
    for i, sig in enumerate([ecg_filtered, art, ppg]):
        plt.subplot(3,1,i+1)
        plt.plot(t[twin], sig[twin], lw=1, c=cp[i])
        sns.despine(bottom=i<2)
        if i<2: plt.xticks([])
        plt.xlim(ta, tb)
    plt.tight_layout()
    plt.show()
```
